```python
import numpy as np
import matplotlib.pyplot as plt

def QPDriver():
    # Define the quadratic objective function
    func = lambda x, y: (x - 1)**2 + (y - 2.5)**2
    gfun = lambda x: np.array([2 * (x[0] - 1), 2 * (x[1] - 2.5)])
    Hfun = lambda x: np.array([[2.0, 0.0], [0.0, 2.0]])

    # Define inequality constraints
    A = np.array([
        [1, -2],
        [-1, -2],
        [-1, 2],
        [1, 0],
        [0, 1]
    ])
    b = np.array([-2.0, -6.0, -2.0, 0.0, 0.0])

    # Initial guess
    x = np.array([2.0, 0.0])
    W = [2,4]

    # Solve using the active set method
    xiter, lm = ASM(x, gfun, Hfun, A, b, W)
    print(xiter)
    # Graphics
    plt.figure(figsize=(8, 8))
    plt.grid()

    # Plot the contours of the objective function
    n = 100
    tx = np.linspace(-1, 5, n)
    ty = np.linspace(-1, 5, n)
    txx, tyy = np.meshgrid(tx, ty)
    ff = func(txx, tyy)
    
    y1 = (tx + 2) / 2  # Constraint 1
    y2 = (6 - tx) / 2  # Constraint 2
    y3 = (tx - 2) / 2  # Constraint 3

    plt.plot(tx, y1, 'r--', label=r'$x - 2y + 2 = 0$')
    plt.plot(tx, y2, 'b--', label=r'$-x - 2y + 6 = 0$')
    plt.plot(tx, y3, 'g--', label=r'$-x + 2y + 2 = 0$')
    plt.axvline(0, color='purple', linestyle='--', label=r'$x = 0$')
    plt.axhline(0, color='orange', linestyle='--', label=r'$y = 0$')

    plt.contour(tx, ty, ff, levels=20, linewidths=1, cmap="viridis")
    feasible_x = [0, 2, 4, 2, 0]
    feasible_y = [0, 0, 1, 2, 1]
    # plt.contourf(*F(x,y),cmap='magma')
    plt.colorbar()
    plt.fill(feasible_x, feasible_y, color='green', alpha=0.5, label='Feasible Region')
    # Plot the feasible region
    # x_vals = np.linspace(-1, 5, 100)
    # plt.fill_between(x_vals, (x_vals - 2) / 2, (6 - x_vals) / 2, where=((x_vals - 2) / 2 >= 0) & ((6 - x_vals) / 2 >= 0), alpha=0.3, color="gray", label="Feasible Region")

    # Plot the solution path
    plt.plot(xiter[0, :], xiter[1, :], marker='.', markersize=10, linestyle='-', linewidth=2, color='r', label="Solution Path")

    plt.xlabel('x', fontsize=16)
    plt.ylabel('y', fontsize=16)
    plt.legend()
    plt.show()


def ASM(x, gfun, Hfun, A, b, W):
    TOL = 1e-10
    dim = len(x)
    g = gfun(x)
    H = Hfun(x)
    iter_max = 1000
    m = A.shape[0]
    
    I = np.arange(m)
    Wc = I.tolist()
    xiter = x[:, np.newaxis]

    for iter in range(iter_max):
        AW = A[W, :] if W else np.empty((0, dim))
        print('W')
        print(W)
        # Ensure H is positive definite
        eigenvalues = np.linalg.eigvalsh(H)
        lam = max(0, -eigenvalues.min() + 1e-10)
        H += lam * np.eye(dim)

        if W:
            M = np.block([[H, -AW.T], [AW, np.zeros((len(W), len(W)))]]);
            RHS = np.concatenate([-g, np.zeros(len(W))])
        else:
            M = H
            RHS = -g
        print('KKT system')
        print(M) 
        print(RHS)
        aux = np.linalg.solve(M, RHS)
        p = aux[:dim]
        lm = aux[dim:] if W else np.array([])
        print('p')
        print(p)
        print('lm')
        print(lm)
        print('x')
        print(x)
        if np.linalg.norm(p) < TOL:
            if W:
                lm = np.linalg.lstsq(AW.T, g, rcond=None)[0]
                if lm.min() >= 0:

                    print(f"A local solution is found at iteration {iter}")
                    print(f"lm = {lm}")
                    break
                else:
                    imin = np.argmin(lm)
                    W.pop(imin)
                    Wc = list(set(I) - set(W))
            else:
                print(f"A local solution is found at iteration {iter}")
                break
        else:
            alp = 1
            Ap = A[Wc, :] @ p
            icand = np.where(Ap < -TOL)[0]
            if icand.size > 0:
                al = (b[Wc][icand] - A[Wc][icand] @ x) / Ap[icand]
                almin = al.min()
                kmin = np.argmin(al)
                alp = min(1, almin)
            x += alp * p
            g = gfun(x)
            H = Hfun(x)
            if alp < 1:
                W.append(Wc[icand[kmin]])
                Wc = list(set(I) - set(W))

        xiter = np.hstack((xiter, x[:, np.newaxis]))

    else:
        print(f"Stopped because the max number of iterations {iter_max} was reached")
    print(x)
    return xiter, lm

QPDriver()
```
