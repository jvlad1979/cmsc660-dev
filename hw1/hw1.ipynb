```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, a=2):
    if x >= 0 and x < 1/2:
        return a*x 
    elif x >= 1/2 and x <= 1:
        return a*(1-x)

def iterate(f, x0, n=100):
    x = [x0]
    for i in range(n):
        x.append(f(x[-1]))
    return x
```

Demos 64 bit floating point iteration of the tent map

```python
fig,ax = plt.subplots(2,1,figsize=(12,10))
x0s = np.random.uniform(0,1,64).astype(np.float64)
for x0 in x0s:
    x = iterate(f,x0)
    ax[0].plot(x,linewidth=.5,color='gray')
ax[0].set_title('Tent Map (64-bit floating)')
ax[0].set_xlabel('Iteration (n)')
ax[0].set_ylabel('f(xn)')
ax[0].hlines([2/3],0,100,linewidth=.5)
x0s = np.random.uniform(0,1,64).astype(np.float32)

for x0 in x0s:
    x = iterate(f,x0)
    ax[1].plot(x,linewidth=.5,color='gray')
ax[1].set_title('Tent Map (32-bit floating)')
ax[1].set_xlabel('Iteration (n)')
ax[1].set_ylabel('f(xn)')
ax[1].hlines([2/3],0,100,linewidth=.5)
```

```python
x0s = np.random.uniform(0,1,64).astype(np.float64)
fig = plt.figure(figsize=(16,8))
for x0 in x0s:
    x = iterate(f,x0)
    plt.plot(x,linewidth=.5)
plt.title('Tent Map (64-bit floating)')
plt.xlabel('Iteration (n)')
plt.hlines([2/3],0,100,linewidth=.5)
plt.show()
```

Demos 32-bit floating point iteration of the tent map

```python
x0s = np.random.uniform(0,1,64).astype(np.float32)
fig = plt.figure(figsize=(16,8))
for x0 in x0s:
    x = iterate(f,x0)
    plt.plot(x,linewidth=.5)
plt.title('Tent Map (32-bit floating)')
plt.xlabel('Iteration (n)')
plt.hlines([2/3],0,100,linewidth=.5)
plt.show()
```

```python
x0 = np.random.uniform(0,1)
fig = plt.figure(figsize=(8,8))
x = iterate(f,x0,n=1000)
plt.scatter(x[:-1],x[1:])
```
