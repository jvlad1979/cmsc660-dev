```python
def f(x, a=2):
    if x >= 0 and x < 1/2:
        return a*x 
    elif x >= 1/2 and x <= 1:
        return a*(1-x)
        # return a - a*x

import numpy as np

# np.linspace(0,1,5)x

# f(.500001)
# f(2/3)
a = 1
b = 2
x = [a/b]
for i in range(100):
    x.append(f(x[-1]))

import matplotlib.pyplot as plt
plt.plot(x)
plt.hlines([(2)/b,(b-1)/b],0,100)
plt.show()
```
