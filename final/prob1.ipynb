```python
from data_utils import get_subset
import numpy as np
import matplotlib.pyplot as plt

X,y=get_subset(labels = [3,8,9])

def plot_clusters(X_transformed,y):
    fig = plt.figure(figsize=(8, 6))
    plt.rc('text', usetex=True)
    plt.rcParams.update({'font.size': 24})
    plt.rc('font', family='serif')
    colors = ['r', 'g', 'b']
    markers = ['o', 's', 'D']

    for i, target in enumerate(np.unique(y)):
        plt.scatter(X_transformed[0,y == target], X_transformed[1,y == target], 
                    color=colors[i], marker=markers[i], alpha=0.1,
                    label=f'Class {target}')

    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.legend(fontsize=18)
    plt.grid()
    # plt.show()
    return fig
```

```python
from linear_algebra import LDA,PCA
```

```python
lda = LDA()
lda.fit(X, y)
X_transformed = lda.transform(X)

fig1a = plot_clusters(X_transformed, y)
fig1a.savefig('./figures/fig1a.svg', transparent=True)
```

```python
pca = PCA()
pca.fit(X)
X_transformed = pca.transform(X)

fig1b = plot_clusters(X_transformed, y)
fig1b.savefig('./figures/fig1b.svg', transparent=True)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

sk_lda =LinearDiscriminantAnalysis()
sk_lda.fit(X.T,y)
X_transformed = sk_lda.transform(X.T).T

plot_clusters(X_transformed, y)
```

```python
from sklearn.decomposition import PCA as skPCA
sk_pca = skPCA(n_components=2)
sk_pca.fit(X.T)
X_transformed = sk_pca.transform(X.T).T

plot_clusters(X_transformed, y)
```
