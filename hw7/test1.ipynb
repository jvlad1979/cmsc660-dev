CG
Input: $x_0$, $r_0 = Ax_0-b$, $p_0 = -r_0$

For k = 0,1,2
    $\alpha_k = -\frac{r_k^Tp_k}{p_k^TAp_k}$
    $x_{k+1} = x_k +\alpha_k p_k$
    $r_{k+1} = r_k +\alpha_k A p_k$
    $\beta_{k+1} = \frac{r_{k+1}Ap_k}{p_kAp_k}$
    $p_{k+1}= -r_{k+1} + \beta_{k+1}p_{k}$

```python
import numpy as np 

N = 69

x0 = np.zeros(N)
b  = np.random.rand(N)

A = np.random.rand(N,N)
A = A.T@A

def conj_grad(A,b,x0):
    r0 = A@x0 -b
    p0 = -r0
    r = r0
    p = p0
    x=x0
    eps = 1e-8
    while np.linalg.norm(r)>eps:
        alpha = -(r.T@r)/(p.T@A@p)
        x =x+ alpha*p
        rold = r
        r = r+ alpha*A@p
        beta = (r.T@r)/(p.T@A@p)
        p = -r + beta*p
    return x
    # r0 = A@x0 -b
    # p0 = -r0 
    # r = r0 
    # p = p0
    # x=x0
    # eps = 1e-8
    # while np.linalg.norm(r)>eps:
    #     alpha = -(r.T@p)/(p.T@A@p)
    #     x =x+ alpha*p
    #     r = r+ alpha*A@p 
    #     beta = (r.T@A@p)/(p.T@A@p)
    #     p = -r + beta*p
    # return x

x = conj_grad(A,b,x0)
```
